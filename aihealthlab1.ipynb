```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Sort correlations with 'TenYearCHD'
correlation_with_TenYearCHD = correlation_matrix['TenYearCHD'].sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=correlation_with_TenYearCHD.index, y=correlation_with_TenYearCHD.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Correlation of Features with TenYearCHD')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Features')
plt.show()
```

```python
high_correlation_features = correlation_with_TenYearCHD[abs(correlation_with_TenYearCHD) > 0.1]
print("Highly correlated features:\n", high_correlation_features)
```

```python
top_4_features = ['age', 'sysBP', 'diaBP', 'glucose']  # Replace with correct column names if different
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, feature in enumerate(top_4_features):
    axes[i].hist(df[feature].dropna(), bins=10, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
x=df.drop('TenYearCHD',axis=1)
y=df['TenYearCHD']
```

```python
x['age_BMI_interaction']=x['age']+x['BMI']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler  # Correct module and class name
scaler = StandardScaler()  # Correct class name

# Fit the scaler on the training data and transform both the training and test data
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)  # Correct method name
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score, classification_report # Import classification_report

# Convert predictions to binary based on a threshold
y_pred_binary = (y_pred >= 0.5).astype(int)

# Calculate various evaluation metrics
accuracy = accuracy_score(y_test, y_pred_binary)
confusion = confusion_matrix(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
precision = precision_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

# Print the results
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Recall:", recall)
print("Precision:", precision)
print("F1 Score:", f1)
report=classification_report(y_test, y_pred_binary) # Now you can use classification_report
print("classification report:\n",report)
```
