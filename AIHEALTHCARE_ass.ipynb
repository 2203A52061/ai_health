```python
import tensorflow as tf

# Define constants
a = tf.constant(10)
b = tf.constant(5)

# Addition
add = tf.add(a, b)

# Subtraction
subtract = tf.subtract(a, b)

# Multiplication
multiply = tf.multiply(a, b)

# Division
divide = tf.divide(a, b)

# Print the results
print("Addition: ", add.numpy())
print("Subtraction: ", subtract.numpy())
print("Multiplication: ", multiply.numpy())
print("Division: ", divide.numpy())
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Function to create and train the model
def train_linear_model(X, Y, title):
# Define the model
    model = tf.keras.Sequential([
       tf.keras.layers.Dense(1, input_shape=(1,), kernel_initializer='zeros')
    ])



# Compile the model
    model.compile(optimizer='sgd', loss='mean_squared_error')

 # Train the model
    print(f"Training model for {title}...")
    model.fit(X, Y, epochs=100, verbose=1)

 # Predict using the model
    Y_pred = model.predict(X)
    print(f"Finished training model for {title}.\n")

    return model, Y_pred

# Generate synthetic data
X = np.linspace(-1, 1, 100)

# Case 1: Y = 3X + 2
Y1 = 3 * X + 2 + np.random.normal(0, 0.1, X.shape)
model1, Y1_pred = train_linear_model(X, Y1, "Y = 3X + 2")

# Case 2: Y = 2X
Y2 = 2 * X + np.random.normal(0, 0.1, X.shape)
model2, Y2_pred = train_linear_model(X, Y2, "Y = 2X")

# Case 3: Y = 2X - 1
Y3 = 2 * X - 1 + np.random.normal(0, 0.1, X.shape)
model3, Y3_pred = train_linear_model(X, Y3, "Y = 2X - 1")

# Plot the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(X, Y1, label='Data')
plt.plot(X, Y1_pred, color='red', label='Fitted line')
plt.title('Y = 3X + 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(X, Y2, label='Data')
plt.plot(X, Y2_pred, color='red', label='Fitted line')
plt.title('Y = 2X')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1, 3, 3)
plt.scatter(X, Y3, label='Data')
plt.plot(X, Y3_pred, color='red', label='Fitted line')
plt.title('Y = 2X - 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.tight_layout()

# Ensure the plot is shown
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)
X1 = np.random.uniform(-1, 1, 100)  # Random values for X1
X2 = np.random.uniform(-1, 1, 100)  # Random values for X2
Y = 3 * X1 + 2 * X2 + 1 + np.random.normal(0, 0.1, X1.shape)  # Y = 3*X1 + 2*X2 + 1 with some noise

# Stack X1 and X2 to form the input features
X = np.stack((X1, X2), axis=1)

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2,), kernel_initializer='zeros')
])

    # Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')

    # Train the model
model.fit(X, Y, epochs=100, verbose=1)

    # Predict using the model
Y_pred = model.predict(X)

    # Plot the results
plt.figure(figsize=(10, 5))

    # Plot actual vs. predicted values
plt.scatter(Y, Y_pred, label='Predicted vs Actual', color='blue')
plt.plot([min(Y), max(Y)], [min(Y), max(Y)], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Actual Y')
plt.ylabel('Predicted Y')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()

    # Print model parameters
weights = model.get_weights()
print("Weights:", weights[0])
print("Bias:", weights[1])
```
